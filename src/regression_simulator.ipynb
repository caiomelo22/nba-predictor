```python
from tqdm import tqdm
import pandas as pd

from helpers.file import save_file
from services.mysql import MySQLService
```

```python
save_pipeline = True
```

```python
mysql = MySQLService()

order_by_clause = "date ASC"
games_df = mysql.get_data("games", order_by_clause=order_by_clause)

season_games_plyrs = mysql.execute_query(f"""
    SELECT g.id as game_id, g.date, g.season, g.is_playoff, g.winner, g.home_id, p.name as player_name, g.away_id, pg.team_id, pg.player_id, pg.minutes, pg.pts, pg.fgm, pg.fga, pg.fg_pct, pg.fg3m, pg.fg3a, pg.fg3_pct, pg.ftm, pg.fta, pg.ft_pct, pg.oreb, pg.dreb, pg.reb, pg.ast, pg.stl, pg.blk, pg.tov, pg.pf, pg.plus_minus
    FROM player_games AS pg
        LEFT JOIN games as g on pg.game_id = g.id 
        LEFT JOIN players as p on pg.player_id = p.id 
    ORDER BY g.date ASC""")

display(games_df)
display(season_games_plyrs)
```

```python
season_games_plyrs.columns
```

```python
stats_columns = ['pts','fgm', 'fga', 'fg_pct', 'fg3m', 'fg3a', 'fg3_pct', 'ftm', 'fta',
    'ft_pct', 'oreb', 'dreb', 'reb', 'ast', 'stl', 'blk', 'tov', 'pf']

for i in tqdm(range(len(stats_columns))):
    stat = stats_columns[i]
    season_games_plyrs[f'{stat}_avg'] = (
        season_games_plyrs.sort_values(by=['player_id', 'date'])
                .groupby('player_id')[stat]
                .transform(lambda x: x.shift(1).rolling(window=5).mean())
    )

display(season_games_plyrs)
```

```python
season_games_plyrs['days_since_last_game'] = (
    season_games_plyrs.sort_values(by=['player_id', 'date'])
             .groupby('player_id')['date']
             .transform(lambda x: x.diff().dt.days.fillna(7))
)

display(season_games_plyrs)
```

```python
season_games_plyrs.columns
```

```python
# Define the opponent_team_id column
season_games_plyrs['opponent_team_id'] = season_games_plyrs.apply(
    lambda row: row['away_id'] if row['team_id'] == row['home_id'] else row['home_id'],
    axis=1
)

# Combine home and away stats into a unified DataFrame for conceded stats
team_stats = pd.concat([
    games_df[['date', 'home_id'] + [f'away_{stat}' for stat in stats_columns]].rename(
        columns={**{'home_id': 'team_id'},
                 **{f'away_{stat}': f'{stat}_conceded' for stat in stats_columns}}
    ),
    games_df[['date', 'away_id'] + [f'home_{stat}' for stat in stats_columns]].rename(
        columns={**{'away_id': 'team_id'},
                 **{f'home_{stat}': f'{stat}_conceded' for stat in stats_columns}}
    )
], ignore_index=True)

# Sort by team_id and date
team_stats = team_stats.sort_values(by=['team_id', 'date'])

# Calculate rolling averages for stats conceded by each team
for stat in stats_columns:
    conceded_stat = f'{stat}_conceded'
    rolling_stat = f'rolling_avg_{conceded_stat}'
    team_stats[rolling_stat] = (
        team_stats.groupby('team_id')[conceded_stat]
        .transform(lambda x: x.shift(1).rolling(window=10, min_periods=1).mean())
    )

# Map opponent stats back to the players DataFrame
# First, merge opponent stats into the players DataFrame using opponent_team_id
rolling_columns = [f'rolling_avg_{stat}_conceded' for stat in stats_columns]
season_games_plyrs = season_games_plyrs.merge(
    team_stats[['team_id', 'date'] + rolling_columns],
    left_on=['opponent_team_id', 'date'],  # Match on opponent and game date
    right_on=['team_id', 'date'],  # Match on team_id and date
    how='left'
)

# Rename columns dynamically for clarity
renamed_columns = {
    f'rolling_avg_{stat}_conceded': f'opponent_avg_conceded_{stat}' for stat in stats_columns
}
season_games_plyrs.rename(columns=renamed_columns, inplace=True)

# Check the final result
display(season_games_plyrs)
```

```python
for col in season_games_plyrs.columns:
    print(col)
```

```python
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error

# Split the dataset into training and testing based on the season
train_data = season_games_plyrs[season_games_plyrs['season'] < 2024]
test_data = season_games_plyrs[season_games_plyrs['season'] == 2024]

# Define feature columns (excluding target stats)
selected_stat = "pts"
additional_stats = ["fg_pct"]
feature_columns = [f"{selected_stat}_avg", f"opponent_avg_conceded_{selected_stat}", "days_since_last_game"]

for add_stat in additional_stats:
    feature_columns.extend([f"{add_stat}_avg", f"opponent_avg_conceded_{add_stat}"])

print("Feature columns:", feature_columns)

# Drop rows with NaN values in feature columns for training and testing
X_train = train_data[feature_columns].dropna()
y_train = train_data.loc[X_train.index, selected_stat]  # Align target with filtered features
X_test = test_data[feature_columns].dropna()
y_test = test_data.loc[X_test.index, selected_stat]  # Align target with filtered features

# Define models to compare
models = {
    "RandomForest": RandomForestRegressor(random_state=42, n_estimators=100),
    "GradientBoosting": GradientBoostingRegressor(random_state=42, n_estimators=100),
    "LinearRegression": LinearRegression(),
    "XGBoost": XGBRegressor(random_state=42, n_estimators=100, use_label_encoder=False, eval_metric='rmse')
}

# Train and evaluate each model
results = {}

best_model_result = 1000
best_model = None

for name, model in models.items():
    print(f"Training {name} model...")
    
    # Train the model
    model.fit(X_train, y_train)
    
    # Predict and evaluate
    y_pred = model.predict(X_test)
    mse_result = mean_squared_error(y_test, y_pred)
    mae_result = mean_absolute_error(y_test, y_pred)
    
    # Store results
    results[name] = {'mse': mse_result, 'mae': mae_result, 'y_test': y_test, 'y_pred': y_pred, 'model': model, "model_name": name}
    
    # Display results
    print(f"{name} - Test MSE: {mse_result:.4f}, Test MAE: {mae_result:.4f}")

    if mae_result < best_model_result:
        best_model_result = mae_result
        best_model = results[name]
```

```python
results_df = pd.DataFrame({
    'Actual': best_model['y_test'].values,
    'Predicted': best_model['y_pred']
})

results_df["Difference"] = results_df.apply(lambda x: x["Actual"] - x["Predicted"], axis=1)
results_df["Negative Diff"] = results_df["Difference"] > 0

display(results_df)

total_pos_diff = results_df["Negative Diff"].sum()
print("Positive difference:", total_pos_diff/len(results_df))
```

```python
# Define confidence interval based on historical model error
confidence_margin = best_model['mae']  # Example: based on standard deviation or MAE of past predictions

# Calculate theoretical bounds
results_df['Lower_Bound'] = results_df['Predicted'] - confidence_margin
results_df['Upper_Bound'] = results_df['Predicted'] + confidence_margin

# Simulate betting strategy
def betting_decision(row, theoretical_line):
    if theoretical_line > row['Upper_Bound']:
        return 'Bet Under'
    elif theoretical_line < row['Lower_Bound']:
        return 'Bet Over'
    else:
        return 'No Bet'

# Apply betting decisions for future games
theoretical_line = 25  # Example: set a constant or dynamic theoretical line
results_df['Bet_Suggestion'] = results_df.apply(lambda row: betting_decision(row, theoretical_line), axis=1)

# Display results
print(results_df[['Actual', 'Predicted', 'Difference', 'Lower_Bound', 'Upper_Bound', 'Bet_Suggestion']])
```
